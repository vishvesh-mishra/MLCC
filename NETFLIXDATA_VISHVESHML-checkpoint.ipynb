```python
#Reading the Data File
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

```

```python
df = pd.read_csv('Netflix Shows.csv', encoding='latin-1')
df.head(100)
```

```python
#REMOVING DUPLICATE ROWS
```

```python
nf = df.drop_duplicates()
nf
```

```python
#DESCRIBING THE DATA FRAME
```

```python
nf.describe()
```

```python
#SORTING THE DATAFRAME IN DESCENDING ORDER OF THE RELEASE YEAR
```

```python
nf.sort_values('release year',ascending=False)
```

```python
#AGGREGATING STATISTICS USING GROUPBY
```

```python
df.groupby(['title']).mean().sort_values('user rating size',ascending=False)
```

```python
#FILTERING THE DATA 
```

```python
new_nf=nf.loc[df['user rating size']>80]
new_nf
```

```python
#DATA VISUALIZATION
```

```python
#BAR GRAPH TO SHOW THE USER RATING SIZE FREQUENCY
```

```python
score_freq = nf['user rating size'].value_counts()

plt.figure(figsize=(12,4))
score_freq.plot.bar(color='purple', alpha=0.5)
plt.xticks(rotation=50)
plt.xlabel("User Rating Size")
plt.ylabel("Frequency of User Rating Size")
plt.title("User Rating Size Frequency")
plt.show()
```

```python
#YEAR IN WHICH THE HIGHEST MOVIES WERE RELEASED
```

```python
print('This is the Year Wise Distribution Of Netflix Shows')
MoviesInEachYear  = df['release year'].value_counts().sort_values(ascending = False)
plt.figure(figsize  = (36,12))
plt.title('Years with the Number of Netlfix Shows')
MoviesInEachYear.plot(kind = 'bar',color='r')
```
